# AI-Driven Health Risk Analysis
This notebook demonstrates how basic AI logic and synthetic data can be used to identify high-risk patient profiles. The goal is to simulate an early-warning dashboard for healthcare intervention.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the synthetic dataset
df = pd.read_csv('synthetic_health_data.csv')
df.head()
```

## Risk Scoring Criteria
Each patient is scored based on the following criteria:
- High Blood Pressure (Systolic > 140 or Diastolic > 90)
- Glucose Level > 140
- BMI > 30 (Obese)
- Age > 60
- Family History of Chronic Illness

A score of 3 or more flags the patient as **High Risk**.

```python
# Distribution of Risk Scores
sns.countplot(data=df, x='Risk_Score', palette='Blues')
plt.title('Risk Score Distribution')
plt.xlabel('Risk Score')
plt.ylabel('Number of Patients')
plt.show()
```

```python
# Visualize High Risk vs Low Risk Patients by Age
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='High_Risk', y='Age', palette='coolwarm')
plt.title('Age Distribution by Risk Category')
plt.xlabel('High Risk (True/False)')
plt.ylabel('Age')
plt.show()
```
